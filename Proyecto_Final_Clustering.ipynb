# K Means Clustering

## Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Import Database

```python
dataset = pd.read_csv('heart.csv')
X = dataset.iloc[:, :-1].values
```

## Encode Values

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
X[:, 1] = le.fit_transform(X[:, 1])
X[:, 2] = le.fit_transform(X[:, 2])
X[:, 6] = le.fit_transform(X[:, 6])
X[:, 8] = le.fit_transform(X[:, 8])
X[:, 10] = le.fit_transform(X[:, 10])

print(X[0])
```

## Model Training

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)
```

## Display Results

```python
print(y_kmeans)
```

## Visualize Clusters

```python
plt.scatter(X[y_kmeans == 0, 4], X[y_kmeans == 0, 6], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 4], X[y_kmeans == 1, 6], s = 100, c = 'orange', label = 'Cluster 2')
plt.scatter(kmeans.cluster_centers_[:, 4], kmeans.cluster_centers_[:, 6], s = 250, c = 'black', label = 'Centroids')
plt.title('Clusters of Heart Disease')
plt.xlabel('Colesterol')
plt.ylabel('Electrocardiograma')
plt.legend()
plt.show()
```
